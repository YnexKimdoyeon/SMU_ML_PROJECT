```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'

file_path = "data/final_data.csv"
df = pd.read_csv(file_path, encoding='utf-8-sig')

# 질병명 컬럼 추출
disease_counts = df['질병명'].value_counts()
```

```python
print(df['질병명'].nunique())
disease_percentages = (df['질병명'].value_counts(normalize=True) * 100).round(2)
print(disease_percentages)
```

```python
#파이차트 그리기
plt.figure(figsize=(10, 10))
plt.pie(disease_counts, labels=disease_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of 질병명')
plt.axis('equal')
plt.show()
```
