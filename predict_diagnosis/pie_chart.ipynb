```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'

file_path = "data/final_data.csv"
df = pd.read_csv(file_path, encoding='utf-8-sig')

# 질병명 컬럼 추출
disease_counts = df['질병명'].value_counts()
```

```python
print(disease_counts)
```

```python
df = df[df['질병명'] != 'non']
main_diseases = disease_counts[disease_counts >= 2000].index
df['질병명'] = df['질병명'].apply(lambda x: x if x in main_diseases else '기타')
print(f"질병분류 갯수: {df['질병명'].nunique()}")
print(df['질병명'].value_counts())

disease_counts = df['질병명'].value_counts()
```

```python
#파이차트 그리기
plt.figure(figsize=(10, 10))
plt.pie(disease_counts, labels=disease_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.show()
```
